```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/sample_data/crime.csv")
```

```python
data.head()
data.info()
data.describe()
```

```python
data.isnull().sum()
```

```python
data
```

```python
from sklearn.impute import SimpleImputer

# Impute missing values in numeric crime-related columns with mean strategy
imputer = SimpleImputer(strategy="mean")
imputed_data = imputer.fit_transform(data[["murder", "gangrape"]])  # Assuming these are numeric

# Update your data with imputed values
data[["murder", "gangrape"]] = imputed_data
```

```python
data
```

```python
import matplotlib.pyplot as plt

specific_value_1 = data["totarea"]
specific_value_2=data['totalcrime']
specific_value_1,specific_value_2
plt.xlabel('total_area')
plt.ylabel('total_crime')
plt.plot(specific_value_1,specific_value_2)
plt.show()
```
